### Grouping and Aggregating - Analyzing and Exploring Your Data
---

```python
# Return the median of the values over the requested axis.
#median age and salary
str(df.loc[:,'Age'].median())+ ' ===> ' +str(df.loc[:,'ConvertedComp'].median())
```

```python
df.median()
```

```python
# Descriptive statistics include those that summarize the central tendency, 
# dispersion and shape of a dataset’s distribution, excluding NaN values.
df.describe()
```

```python
cntry_grp = df.groupby(['Country'])
```

```python
cntry_grp.get_group('Japan')
```

```python
cntry_grp['ConvertedComp'].median().loc['India']
```

```python
cntry_grp['ConvertedComp'].agg(['mean','median']).loc['India']
```

```python
# % of people that have worked with python from different countries
cntry_grp['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').value_counts(normalize=True)).loc['India']
```
